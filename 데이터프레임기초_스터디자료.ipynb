```python
from PublicDataReader import Kbland
api = Kbland()
params = {
    "월간주간구분코드": "02", # 01: 월간, 02: 주간
    "매물종별구분": "01",    # 01: 아파트, 08: 연립, 09: 단독, 98: 주택종합
    "매매전세코드": "01",    # 01: 매매, 02: 전세
    "지역코드": "11",       # 11: 서울, 42: 경기
    "기간": "1",           # 최근 1년: 1, 최근 2년: 2 (미지정 시 전체 기간으로 조회)
}

df = api.get_price_index(**params)
df
```

<h3>13. df[df[컬럼명] == 조건명] <br/>
df[df[컬럼명] > 조건명], df[df[컬럼명] >= 조건명] <br/>
df[df[컬럼명] < 조건명], df[df[컬럼명] <= 조건명]
</h3>

> 데이터프레임에 조건을 걸어 조건에 만족하는 행만 볼 수 있습니다.

∙ df[df[컬럼명] == 조건명]

```python
df[df['지역명'] == '용산구'] # 지역명이 용산구인 데이터만 보여집니다.
```

∙ df[df[컬럼명] >= 조건명]

```python
df[df['가격지수'] >= 95] # 가격지수가 95 이상인 행만 가져옵니다.
```

∙ df[df[컬럼명] <= 조건명]

```python
df[df['가격지수'] <= 85] # 가격지수가 85이하인 행만 가져옵니다.
```
