```python
import os
import json
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import matplotlib.font_manager as fm

# 폰트 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fontprop = fm.FontProperties(fname=font_path, size=12)

folder_path = "/content/collect_jobKorea"
start_day = 1
end_day = 31

# 각 날짜별로 파일을 읽어서 데이터를 수집
data_by_day = {}

# month 세팅
start_month = 11
end_month = 12

# 결과 담을 곳
results = []

for month in range(start_month, end_month + 1):
  for day in range(start_day, end_day + 1):
    day = str(day).zfill(2)
    file_path = os.path.join(folder_path, f'2023-{month}-{day}_job_counting.txt')

    if not os.path.exists(file_path):
        continue

    with open(file_path, 'r') as file:
        lines = file.readlines()

        # 파일 내의 각 줄을 개별적으로 JSON으로 파싱하여 저장
        daily_data = [json.loads(line.strip().replace("'", '"')) for line in lines]
        result = {
            'day': f'2023-{month}-{day}',
            'value': daily_data
        }

        results.append(result)

print(len(results))
print(results)

# 각 날짜별로 subName의 giCnt를 추출하여 저장
data_by_subName = {}
days = []

for result in results:
    day = result['day']
    days.append(day)
    value = result['value']

    # subNamee별 giCnt를 저장할 딕셔너리 초기화
    for item in value:
        subName = item['subName']
        gi_cnt = item['giCnt']

        if subName not in data_by_subName:
            data_by_subName[subName] = []

        data_by_subName[subName].append(gi_cnt)

# 그래프 크기 조절
plt.figure(figsize=(20, 15))

# 각 subName의 giCnt 시계열 추이 시각화
for subName, gi_cnt_values in data_by_subName.items():
    plt.plot(days, gi_cnt_values, label=f'{subName}')

plt.xlabel('날짜', fontproperties=fontprop)
plt.ylabel('일자리수', fontproperties=fontprop)
plt.title(f"{start_month}~{end_month}월 IT 일자리 추이", fontproperties=fontprop)
plt.legend(prop=fontprop, loc='center')
plt.show()
```
